# Breast Cancer Detection (Wisconsin Dataset)

A complete, fast, and well-explained end-to-end workflow for classifying malignant vs benign tumors using scikit-learn's Wisconsin Breast Cancer dataset. The notebook covers data audit, modeling (Logistic Regression, SVM, Random Forest), cross-validation and tuning, evaluation (ROC/PR curves, calibration), persistence, and interpretability (feature importance, SHAP).

```python
# 1) Import Required Libraries
import os, sys, warnings, math, json, time, random
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display

from sklearn import datasets
from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV, RandomizedSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    classification_report, confusion_matrix, roc_curve, auc,
    precision_recall_curve, average_precision_score, brier_score_loss
)
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.calibration import CalibratedClassifierCV, calibration_curve

import joblib

# Optional SHAP
try:
    import shap
    SHAP_AVAILABLE = True
except Exception:
    SHAP_AVAILABLE = False

# Reproducibility
RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)
random.seed(RANDOM_STATE)

plt.style.use('seaborn-v0_8')
plt.rcParams['figure.figsize'] = (10, 6)
plt.rcParams['axes.grid'] = True
print('Environment ready. SHAP available:', SHAP_AVAILABLE)

# 2) Load Breast Cancer Dataset (scikit-learn)
sk_data = datasets.load_breast_cancer()
X = pd.DataFrame(sk_data.data, columns=sk_data.feature_names)
y = pd.Series(sk_data.target, name='target')
class_names = list(sk_data.target_names)
print('Data shape:', X.shape, 'Target shape:', y.shape)
print('Classes:', class_names)

# 3) Quick Data Audit and EDA
print('\nHead:')
display(X.head())
print('\nDescribe:')
display(X.describe().T)
print('Missing values per column:')
display(X.isna().sum())

# Class balance
class_counts = y.value_counts().sort_index()
plt.figure()
sns.barplot(x=class_counts.index.map({0:'malignant',1:'benign'}), y=class_counts.values)
plt.title('Class Distribution')
plt.ylabel('Count')
plt.xlabel('Class')
plt.show()

# Correlation heatmap (top features to keep it fast)
plt.figure(figsize=(12, 10))
sns.heatmap(X.corr(), cmap='coolwarm', center=0)
plt.title('Feature Correlation Heatmap')
plt.show()

# 4) Train/Validation/Test Split with Stratification
# 80/20 -> train_temp/test, then 75/25 of train_temp -> train/val
X_train_temp, X_test, y_train_temp, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=RANDOM_STATE
)
X_train, X_val, y_train, y_val = train_test_split(
    X_train_temp, y_train_temp, test_size=0.25, stratify=y_train_temp, random_state=RANDOM_STATE
)  # 0.25 of 0.8 = 0.2; final: 60/20/20

print(f'Train/Val/Test sizes: {X_train.shape[0]}/{X_val.shape[0]}/{X_test.shape[0]}')

# Verify stratification ratios
for name, yy in [('all', y), ('train', y_train), ('val', y_val), ('test', y_test)]:
    ratio = yy.mean()
    print(f'{name:>5} positive ratio (benign=1): {ratio:.3f}')

# 5) Preprocessing Pipeline (Imputation + Scaling)
preprocess = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

# Fit on train only
preprocess.fit(X_train)
X_train_p = preprocess.transform(X_train)
X_val_p = preprocess.transform(X_val)
X_test_p = preprocess.transform(X_test)

# 6) Baseline Model: Logistic Regression
log_reg = Pipeline([
    ('preprocess', preprocess),
    ('clf', LogisticRegression(max_iter=1000, random_state=RANDOM_STATE))
])
log_reg.fit(X_train, y_train)
val_pred_lr = log_reg.predict(X_val)
val_proba_lr = log_reg.predict_proba(X_val)[:, 1]

print('\nBaseline Logistic Regression (Validation):')
print('Accuracy:', accuracy_score(y_val, val_pred_lr))
print('Precision:', precision_score(y_val, val_pred_lr))
print('Recall:', recall_score(y_val, val_pred_lr))
print('F1:', f1_score(y_val, val_pred_lr))
print('\nClassification Report:\n', classification_report(y_val, val_pred_lr, target_names=class_names))

cm = confusion_matrix(y_val, val_pred_lr)
plt.figure()
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
plt.title('Confusion Matrix - Logistic Regression (Val)')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()

# 7) Additional Models: SVM and Random Forest
svc = Pipeline([
    ('preprocess', preprocess),
    ('clf', SVC(probability=True, random_state=RANDOM_STATE))
])
rf = Pipeline([
    ('preprocess', preprocess),
    ('clf', RandomForestClassifier(n_estimators=300, random_state=RANDOM_STATE))
])

svc.fit(X_train, y_train)
rf.fit(X_train, y_train)

val_pred_svc = svc.predict(X_val)
val_proba_svc = svc.predict_proba(X_val)[:, 1]
val_pred_rf = rf.predict(X_val)
val_proba_rf = rf.predict_proba(X_val)[:, 1]

print('\nSVC (Validation): Acc=', accuracy_score(y_val, val_pred_svc), 'F1=', f1_score(y_val, val_pred_svc))
print('RandomForest (Validation): Acc=', accuracy_score(y_val, val_pred_rf), 'F1=', f1_score(y_val, val_pred_rf))

# 8) Cross-Validation and Hyperparameter Tuning
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)

param_lr = {'clf__C': [0.01, 0.1, 1, 10, 100], 'clf__penalty': ['l2'], 'clf__solver': ['lbfgs']}
param_svc = {'clf__C': [0.1, 1, 10, 100], 'clf__kernel': ['rbf', 'linear'], 'clf__gamma': ['scale', 'auto']}
param_rf = {'clf__n_estimators': [200, 400, 800], 'clf__max_depth': [None, 5, 10, 20]}

# To keep runtime reasonable, we do quick GridSearch on LR and SVC, RandomizedSearch on RF
lr_grid = GridSearchCV(Pipeline([('preprocess', preprocess), ('clf', LogisticRegression(max_iter=2000, random_state=RANDOM_STATE))]),
                      param_grid=param_lr, scoring='f1', cv=cv, n_jobs=-1)
svc_grid = GridSearchCV(Pipeline([('preprocess', preprocess), ('clf', SVC(probability=True, random_state=RANDOM_STATE))]),
                       param_grid=param_svc, scoring='f1', cv=cv, n_jobs=-1)
rf_rand = RandomizedSearchCV(Pipeline([('preprocess', preprocess), ('clf', RandomForestClassifier(random_state=RANDOM_STATE))]),
                            param_distributions={
                                'clf__n_estimators': [200, 300, 400, 600, 800, 1000],
                                'clf__max_depth': [None, 4, 6, 8, 10, 12, 16]
                            }, n_iter=12, scoring='f1', cv=cv, n_jobs=-1, random_state=RANDOM_STATE)

print('\nRunning hyperparameter searches (this is quick but may take ~10-30s)...')
lr_grid.fit(X_train, y_train)
svc_grid.fit(X_train, y_train)
rf_rand.fit(X_train, y_train)

print('Best LR:', lr_grid.best_params_, 'CV F1:', lr_grid.best_score_)
print('Best SVC:', svc_grid.best_params_, 'CV F1:', svc_grid.best_score_)
print('Best RF:', rf_rand.best_params_, 'CV F1:', rf_rand.best_score_)

# Compare on validation
candidates = {
    'LR': lr_grid.best_estimator_,
    'SVC': svc_grid.best_estimator_,
    'RF': rf_rand.best_estimator_
}

val_scores = {}
for name, est in candidates.items():
    y_pred = est.predict(X_val)
    y_proba = est.predict_proba(X_val)[:, 1]
    fpr, tpr, _ = roc_curve(y_val, y_proba)
    val_scores[name] = {
        'acc': accuracy_score(y_val, y_pred),
        'precision': precision_score(y_val, y_pred),
        'recall': recall_score(y_val, y_pred),
        'f1': f1_score(y_val, y_pred),
        'auroc': auc(fpr, tpr)
    }

print('\nValidation scores (tuned models):')
for k, v in val_scores.items():
    print(k, v)

# Select best model by F1 on validation
best_name = max(val_scores, key=lambda k: val_scores[k]['f1'])
best_model = candidates[best_name]
print(f'\nSelected best model: {best_name}')

# 9) Evaluation: Confusion Matrix, ROC-AUC, PR Curve (Val)
y_val_pred = best_model.predict(X_val)
y_val_proba = best_model.predict_proba(X_val)[:, 1]

print('\nValidation Metrics (Best Model):')
print('Accuracy:', accuracy_score(y_val, y_val_pred))
print('Precision:', precision_score(y_val, y_val_pred))
print('Recall:', recall_score(y_val, y_val_pred))
print('F1:', f1_score(y_val, y_val_pred))
print('\nClassification Report:\n', classification_report(y_val, y_val_pred, target_names=class_names))

cm = confusion_matrix(y_val, y_val_pred)
plt.figure()
sns.heatmap(cm, annot=True, fmt='d', cmap='Purples', xticklabels=class_names, yticklabels=class_names)
plt.title(f'Confusion Matrix - {best_name} (Val)')
plt.xlabel('Predicted'); plt.ylabel('True')
plt.show()

# ROC curve
fpr, tpr, _ = roc_curve(y_val, y_val_proba)
roc_auc = auc(fpr, tpr)
plt.figure()
plt.plot(fpr, tpr, label=f'ROC curve (AUC={roc_auc:.3f})')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate'); plt.ylabel('True Positive Rate')
plt.title(f'ROC - {best_name} (Val)')
plt.legend(); plt.show()

# Precision-Recall
prec, rec, _ = precision_recall_curve(y_val, y_val_proba)
ap = average_precision_score(y_val, y_val_proba)
plt.figure()
plt.plot(rec, prec, label=f'PR curve (AP={ap:.3f})')
plt.xlabel('Recall'); plt.ylabel('Precision')
plt.title(f'Precision-Recall - {best_name} (Val)')
plt.legend(); plt.show()

# 10) Feature Importance and Coefficients
if hasattr(best_model.named_steps['clf'], 'feature_importances_'):
    importances = best_model.named_steps['clf'].feature_importances_
    feat_imp = pd.Series(importances, index=X.columns).sort_values(ascending=False)
    plt.figure(figsize=(10,8))
    sns.barplot(x=feat_imp.values[:15], y=feat_imp.index[:15])
    plt.title(f'Top 15 Feature Importances ({best_name})')
    plt.show()
elif hasattr(best_model.named_steps['clf'], 'coef_'):
    coefs = best_model.named_steps['clf'].coef_[0]
    coef_series = pd.Series(coefs, index=X.columns).sort_values()
    plt.figure(figsize=(10,10))
    coef_series.tail(15).plot(kind='barh', color='g')
    coef_series.head(15).plot(kind='barh', color='r')
    plt.title(f'Top +/- Coefficients ({best_name})')
    plt.show()
else:
    print('Model does not expose feature importances or coefficients.')

# 11) Calibration of Probabilities
uncal_brier = brier_score_loss(y_val, y_val_proba)
print('Uncalibrated Brier score:', uncal_brier)

calibrator = CalibratedClassifierCV(best_model.named_steps['clf'], method='sigmoid', cv=cv)
calibrated = Pipeline([('preprocess', preprocess), ('clf', calibrator)])
calibrated.fit(X_train, y_train)

y_val_proba_cal = calibrated.predict_proba(X_val)[:, 1]
cal_brier = brier_score_loss(y_val, y_val_proba_cal)
print('Calibrated Brier score:', cal_brier)

prob_true, prob_pred = calibration_curve(y_val, y_val_proba, n_bins=10)
prob_true_cal, prob_pred_cal = calibration_curve(y_val, y_val_proba_cal, n_bins=10)
plt.figure()
plt.plot(prob_pred, prob_true, marker='o', label='Uncalibrated')
plt.plot(prob_pred_cal, prob_true_cal, marker='o', label='Calibrated')
plt.plot([0,1],[0,1],'k--')
plt.xlabel('Mean predicted probability'); plt.ylabel('Fraction of positives')
plt.title('Calibration Curve (Val)')
plt.legend(); plt.show()

# 12) End-to-End Pipeline and Model Persistence
# Refit best base estimator on train+val
X_trval = pd.concat([X_train, X_val], axis=0)
y_trval = pd.concat([y_train, y_val], axis=0)

best_model.fit(X_trval, y_trval)

# Evaluate on test
y_test_pred = best_model.predict(X_test)
y_test_proba = best_model.predict_proba(X_test)[:, 1]
print('\nTest Metrics (Best Model):')
print('Accuracy:', accuracy_score(y_test, y_test_pred))
print('Precision:', precision_score(y_test, y_test_pred))
print('Recall:', recall_score(y_test, y_test_pred))
print('F1:', f1_score(y_test, y_test_pred))
fpr_t, tpr_t, _ = roc_curve(y_test, y_test_proba)
print('ROC-AUC:', auc(fpr_t, tpr_t))

# Save
os.makedirs('../models', exist_ok=True)
model_path = '../models/breast_cancer_clf.joblib'
joblib.dump(best_model, model_path)
print('Saved model to', model_path)

# Load and re-verify
loaded = joblib.load(model_path)
check_pred = loaded.predict(X_test[:5])
print('Loaded model sanity predictions:', check_pred.tolist())

# 13) Model Interpretability with SHAP
if SHAP_AVAILABLE:
    try:
        # Choose explainer by model type
        base_clf = best_model.named_steps['clf']
        if isinstance(base_clf, RandomForestClassifier):
            explainer = shap.TreeExplainer(best_model.named_steps['clf'])
            shap_values = explainer.shap_values(preprocess.transform(X_val))
            # For binary, TreeExplainer returns list [class0, class1]
            shap.summary_plot(shap_values[1], preprocess.transform(X_val), feature_names=X.columns, show=False)
            plt.title('SHAP Summary (RF)')
            plt.show()
        elif isinstance(base_clf, LogisticRegression):
            explainer = shap.LinearExplainer(best_model.named_steps['clf'], preprocess.transform(X_train), feature_dependence='independent')
            shap_vals = explainer.shap_values(preprocess.transform(X_val))
            shap.summary_plot(shap_vals, preprocess.transform(X_val), feature_names=X.columns, show=False)
            plt.title('SHAP Summary (LR)')
            plt.show()
        else:
            explainer = shap.KernelExplainer(lambda d: best_model.predict_proba(pd.DataFrame(preprocess.inverse_transform(d), columns=X.columns))[:,1], preprocess.transform(X_train[:100]))
            shap_vals = explainer.shap_values(preprocess.transform(X_val[:50]), nsamples=100)
            shap.summary_plot(shap_vals, preprocess.transform(X_val[:50]), feature_names=X.columns, show=False)
            plt.title('SHAP Summary (Kernel)')
            plt.show()
    except Exception as e:
        print('SHAP failed, skipping interpretability:', e)
else:
    print('SHAP not available; skipping.')

# 14) Batch Inference and Error Analysis
results_df = pd.DataFrame({
    'y_true': y_test.values,
    'y_pred': y_test_pred,
    'proba': y_test_proba
})
results_df['correct'] = (results_df['y_true'] == results_df['y_pred'])

plt.figure()
plt.hist(results_df.loc[results_df['y_true']==1, 'proba'], bins=20, alpha=0.6, label='Benign (1)')
plt.hist(results_df.loc[results_df['y_true']==0, 'proba'], bins=20, alpha=0.6, label='Malignant (0)')
plt.xlabel('Predicted probability (class=1)'); plt.ylabel('Count')
plt.title('Probability Distributions by True Class (Test)')
plt.legend(); plt.show()

# Top errors
fp = results_df[(results_df.y_true==0) & (results_df.y_pred==1)].sort_values('proba', ascending=False).head(5)
fn = results_df[(results_df.y_true==1) & (results_df.y_pred==0)].sort_values('proba', ascending=True).head(5)
print('\nTop 5 False Positives:\n', fp)
print('\nTop 5 False Negatives:\n', fn)

# 15) Unit Tests for Data and Pipeline
print('\nRunning quick unit tests...')

# No NaNs after preprocessing
assert not np.isnan(preprocess.transform(X_train)).any(), 'NaNs found after preprocessing on train.'

# Feature count consistent
assert preprocess.transform(X_train).shape[1] == X.shape[1], 'Feature count mismatch after preprocessing.'

# Stratification ratios within tolerance
base_ratio = y.mean()
for name, yy in [('train', y_train), ('val', y_val), ('test', y_test)]:
    assert abs(yy.mean() - base_ratio) < 0.05, f'Stratification drift in {name}'

# predict_proba sums to 1
pp = best_model.predict_proba(X_test[:10])
assert np.allclose(pp.sum(axis=1), 1.0, atol=1e-6), 'predict_proba rows do not sum to 1.'

print('All tests passed.')
```
